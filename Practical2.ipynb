```python
import pandas as pd

data = {
    "Rooms": [2,3,4,3,5],
    "Trees": [1,2,3,2,4],
    "Type":["Apartment","Villa","Bungalow","Apartmet0","Villa"]
}
df = pd.DataFrame(data)
print("Original DataFrame")
print(df)
```

```python
#Normalization by min max
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['Rooms','Trees']] = scaler.fit_transform(df[['Rooms','Trees']])
print("\nDAtaFrame after Min Max Normalization")
print(df)
```

```python
import pandas as pd

data = {
    "Rooms": [2,3,4,3,5],
    "Trees": [1,2,3,2,4],
    "Type":["Apartment","Villa","Bungalow","Apartmet0","Villa"]
}
df_original= pd.DataFrame(data)

print("Original DataFrame")
print(df_original)

bins = [0,2.5,3.5,5.5]
labels = ['small','medium','large']
df_original[' Room_category '] = pd.cut(df_original['Rooms'],bins=bins,labels=labels,right=True)
df_original[' Trees_category '] = pd.cut(df_original['Trees'],bins=bins,labels=labels,right=True)

print("\nDataFrame after Discretization")
print(df_original)

```
